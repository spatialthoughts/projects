```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Indian_Premier_League_seasons_and_results'
```

```python
dfs = pd.read_html(url)
```

Check number of tables on the page.

```python
len(dfs)
```

Extract the required table.

```python
teams = dfs[4]
teams
```
