## Filter a Shapefile using an Excel Sheet

This notebook shows how to read data from an Excel sheet containing IDs that need to be selected from another shapefile. We used Pandas `isin()` function to apply a filter for the ids from the Excel file.

```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
roads = gpd.read_file('all_roads.shp')
roads
```

```python
roads['osm_id'] = roads['osm_id'].astype(np.int64)
```

```python
subset = pd.read_excel('subset.xlsx')
subset
```

```python
id_list = subset['osm_id']
```

```python
roads[roads['osm_id'].isin(id_list.values)]
```
